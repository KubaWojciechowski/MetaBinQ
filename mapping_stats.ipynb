```python
import numpy as np
import matplotlib.pyplot as plt
from Bio import SeqIO
```

```python
def long_contigs_ids(fasta_file, l):
    """Return list of ids of contigs longer than l
    
    Inputs: 
    fasta_file - Fasta file with contigs
    l - minimal length of contig
    
    Output: list of ids of contigs longer than l
    """
    
    contigs_ids = []
    for record in SeqIO.parse(fasta_file, "fasta"):
        if len(record.seq) > l:
            contigs_ids.append(record.id)
    
    return contigs_ids
```
